```python
%matplotlib inline
import cv2
from matplotlib import pyplot as plt
import numpy as np
import glob
```

```python
image = cv2.imread("media/test_images/straight_lines1.jpg")
```

## Visualization part

```python
def show_image(image, title='Image', cmap_type='gray'):
    plt.imshow(image, cmap = cmap_type)
    plt.title(title)
```

## Read In An Image

```python
# Reading an image
images_paths = glob.glob("media/test_images/*.jpg")
images = [plt.imread(image) for image in images_paths]

# getting img dimensions and printing it
image_idx = 0

print("This image is of shape {}".format(images[image_idx].shape))
show_image(images[image_idx], "Image number one")
```

```python
# so that changes are done to the copy
lane_img=np.copy(images[image_idx])
```

```python
def canny(image):
     #making the image gray to reduce processing time
    gray_img= cv2.cvtColor(lane_img, cv2.COLOR_RGB2GRAY)
    
    #reduce noise to reduce false edges
#gaussian blur smoothening

    blur_img =cv2.GaussianBlur(gray_img, (5,5), 0)
    
    # canny to find edges
#derivative small deriv small change big deriv is an edge
    canny_img = cv2.Canny(blur_img, 50, 150)
    return canny_img
```

```python
canny_img = canny(lane_img)
```

```python
print("This image is of shape {}".format(images[image_idx].shape))
show_image(canny_img, "Image number one")
```

```python
#detect lanes using hough

def lane(image):
    height = image.shape[0]
    polygons= np.array([
        [(200,height), (1100,height),(630,410)]
    ])
    mask = np.zeros_like(image)
    cv2.fillPoly(mask,polygons,255)
    #bitwise to show only whats in the mask
    masked_img= cv2.bitwise_and(image, mask)
    
    
    return masked_img
```

```python
# bitwise and operation mask and canny
cropped_img = lane(canny_img)
print("This image is of shape {}".format(images[image_idx].shape))
show_image(cropped_img, "Image number one")
```

```python
def display_lines(img,lines):
    line_image= np.zeros_like(image)
    if lines is not None:
        for line in lines:
            x1, y1, x2, y2 = line.reshape(4)
            cv2.line(line_image, (x1,y1), (x2,y2), (255, 0,0), 10)
            
    return line_image
```

```python
#hough for straight lines
#sufficiently small bins

lines= cv2.HoughLinesP(cropped_img,2, np.pi/180, 100, np.array([]), minLineLength=40, maxLineGap=5)
line_img= display_lines(lane_img, lines)
print("This image is of shape {}".format(images[image_idx].shape))
show_image(line_img, "Image number one")
```
